## Lists
A list is a sequenced collection of different objects such as integers, strings, and even other lists as well. The address of each element within a list is called an <b>index</b>. An index is used to access and refer to items within a list.

```python
# Create a list
myList = ['Eagles', 'Their Greatest Hits', 1976, '00:43:08', 'Rock', 32.2, 42, '17-Feb-76', 9.5]
```

The indices start from **0** just like the strings.

```python
# Accessing elements of a list
print('The second element at index=1 is : ', myList[1])
print('The second from the last element at index -2 is : ', myList[-2]) 
```

```python
# Nested Lists
myList_nested = [('Michael Jackson', 'Eagles'), [10.0, 9.0], (65, 41)]
print(myList_nested[0][1])
```

### List Operations

```python
# List slicing
print(myList[1:5])
```

```python
# Append (to add a single element/argument as is)
myList.append(['Songs', 'English'])
print(myList)
```

```python
# Extend (to add elements to the list)
myList.extend(['Popular', 'English Singers'])
print(myList)
```

**Unlike strings lists are mutable**

```python
# Change element of a string
print('My list before mutation :')
print(myList)
myList[0] = 'Pink Floyd'
print('My list after mutation :')
print(myList)
```

### Converting strings to lists using .split()
A string can be converted to a list using <code>split</code>.  For example, the method <code>split</code> translates every group of characters separated by a space into an element in a list.

```python
# Splitting a string using .split method, by default 'space' is the splitting criteria
myString = "Manchester is a beautiful city"
myString_Splitted = myString.split()
print(myString_Splitted)
```

### Copy and Clone List
When we set one variable <b>B</b> equal to <b>A</b>, both <b>A</b> and <b>B</b> are referencing the same list in memory:

```python
# Copy (copy by reference) the list A
A = ["rock", 8.8, 1.4]
B = A
print('A: ', A)
print('B: ', B)
```

<font color='red'>Initially, the value of the first element in <b>B</b> is set as "rock". If we change the first element in <b>A</b> to <b>"banana"</b>, we get an unexpected side effect.  As <b>A</b> and <b>B</b> are referencing the same list, if we change list <b>A</b>, then list <b>B</b> also changes. If we check the first element of <b>B</b> we get "banana" instead of "rock":</color>

```python
print('B[0]: ', B[0])
A[0] = "banana"
print('B[0]: ', B[0])
```

**To avoid these problems downstream, either clone the '<u>list by value</u>' aka '<u>shallow copy</u>'**

```python
B = A[:]
print('B[0]: ', B[0])
A[0] = "banana"
print('B[0]: ', B[0])
```

### Deep Copy
The catch with the 'shallow copy' is that in case of a nested list, although the primary elements (top level elements) of a list might be preserved but if an element is nested list then its elements still will be prone to changes. The solution in such cases is **<u>deep copy</u>**.

```python
# Deep copy example
C = [A, 'alpha', '10', 25.4]
print(C)
```

```python
# Demonstration of the problem
A[0] = 'apple'
print(C)
```

```python
# The solution
import copy as copy
D = copy.deepcopy(C)
print(D)
A[0] = 'mango'
print(D)             # Verify that D remains unchanged
```

## <font color='grey'> Chapter 2: Dictionaries</font>
A dictionary consists of keys and values. It is helpful to compare a dictionary to a list. Instead of being indexed numerically like a list, dictionaries have keys. These keys are the keys that are used to access values within a dictionary.  

The best example of a dictionary can be accessing person's detais using the **social security number**.   
Here the social security number which is a unique number will be the **key** and the details of the people will be the **values** associated with it.

### Creating a Dictionary and Accessing its Elements

```python
# Creating a dictionary
myDict = {'key1': 6, 'key2': "Caesar", 'key3': '43', 'key4': 54.1, 'key5': (1,5,6,3), 'key6':[1,'A', '5', 1.5]}
print(myDict)
```

Each key is separated from its value by a colon "<code>:</code>".  Commas separate the items, and the whole dictionary is enclosed in curly braces. An empty dictionary without any items is written with just two curly braces, like this  "<code>{}</code>".

**Keys can only be strings, numbers, or tuples, but values can be any data type.**

```python
# Accessing the values by the key
print(myDict['key2'])
print(myDict['key5'])
```

```python
# Get all keys in a dictionary
keys_lst = myDict.keys()
print(keys_lst)
print(type(keys_lst))      # note that this is not a list, it's 'dict_keys' class
```

```python
# Get all the values in a dictionary
values_lst = myDict.values()
print(values_lst)
print(type(values_lst))    # note that this is not a list, it's 'dict_values' class
```

### Editing, Adding and Removing Elements from a Dictionary

```python
# Edit the value of a key
print(myDict['key1'])
myDict['key1'] = 'Rome'
print(myDict['key1'])
```

```python
# Append a value/key pair into a dictionary
myDict['country'] = 'Italy'
print(myDict)
print(myDict['country'])
```

## <font color='grey'> Chapter 3: Sets </font>

Sets in Python are **<u>unordered collections</u>** of **<u>unique elements</u>**. They are mutable containers built using curly braces {} or the set() constructor, and they are primarily used for eliminating duplicates and performing mathematical set operations like unions and intersections.

### Creating a set

```python
# Create a set
mySet = {"Birmingham", "Manchester", "Glasgow", "Liverpool", "London", "Manchester", "Edinburgh", "London"}
print(mySet)
```

**A list can also be converted to a set.**

```python
# Converting list to a set
myList = ["A", "B", "C", "D", "A", "C", "E"]
convSet = set(myList)
print(type(convSet))
print(convSet)
```

### Set Operations

```python
# Add an element to a set
print(mySet)
mySet.add("Newcastle")
print(mySet)
```

```python
# Set is a collection of unique elements so adding same elements twice will not modify the set
mySet.add("Newcastle")
print(mySet)
```

```python
# Removing an element from a set
mySet.remove("Liverpool")
print(mySet)
```

### Logic Operations on Sets

```python
# Example sets
setA = {"A", "C", "D", "B", "E", "C", "B"}
setB = {"C", "F", "G", "E", "D", "H", "I"}
print(setA)
print(setB)
```

```python
# Intersection of sets
intersection_AB = setA & setB
print(intersection_AB)
# Alternatively
inters_AB = setA.intersection(setB)
print(inters_AB)
```

```python
# Difference between setA and setB
diff_AB = setA.difference(setB)  #Elements present in setA but not in setB
print(diff_AB)
diff_BA = setB.difference(setA)  #Elements present in setB but not in setA
print(diff_BA)
```

```python
# Union of sets
union_AB = setA.union(setB)
print(union_AB)
```

```python
# Check if a set is a superset of another
isSuperset = set(setA).issuperset(setB)
print(isSuperset)
```

```python
# Check if a set is a subset of another
ifSubset = set(setA).issubset(setB)
print(ifSubset)
```
