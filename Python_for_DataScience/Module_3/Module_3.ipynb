### Branching
Branching in programming is a control flow mechanism where the computer chooses between two or more alternative paths of execution based on whether a specific condition is true or false.
Instead of executing instructions in a straight line from top to bottom, the program "branches" out into different directions, allowing the code to make decisions dynamically during runtime.  

```python
# If statements
cutoff = 95

score = 90

if score > cutoff:
    print("Congratulations! You have passed the exam.")

print("Please input the score of next candidate")
```

```python
# Else (+ if) statements
cutoff = 95

score = 90

if score > cutoff:
    print("Congratulations! You have passed the exam.")

else:
    print("Better luck next time.")

print("Please input the score of next candidate")
```

```python
# Elif statements
cutoff = 80
best_scorers = 90

score = 82

if score > best_scorers:
    print("Congratulations! You are one of the best scorers.")
elif score > cutoff and score < best_scorers:               # You'll find more details about the logical operator 'and' in later sections
    print("Congratulations! You passed the exam.")
else:
    print("Sorry, better luck next time.")
```

## <font color="grey">Chapter 2: Loops</font>
A loop is a programming tool that repeats a block of code over and over again so you don't have to write the same lines manually.  

Instead of writing a line of code 10 times, you write it once inside a loop and tell Python how many times to repeat it.

### Range
A range loop is a specific type of <code>for</code> loop (you'll learn more about this in future sections) that repeats a block of code a set number of times using the built-in <code>range()</code> function.

```python
# Example of range
range(15)
```

### <code>for</code> Loop

A <code>for</code> loop is a programming tool used to repeat a block of code a specific number of times, or to step through a collection of items one by one.  

In simple terms, you use a for loop when you already know how many times you need to repeat an action, or how many items you need to process.

```python
# Example of for loop
for x in range(11):
    print(2*x)
```

```python
# Looping through elements of a list and its indices
myList = ["A", "B", "C", "D", "E", "F", "E"]

for i, char in enumerate(myList):
    print(i, char)
```

### <code>while</code> Loop

A while loop is a programming tool used to repeat a block of code over and over again as long as a specific condition remains true.

Unlike a for loop (which usually counts a set number of times), a while loop is based entirely on a condition. You use it when you don't know exactly how many repetitions it will take to finish the job.

```python
# Example of while loop
threshold = 10
count = 0

while count < threshold:
    print(count)
    count = count+1

print("I am out of 'while' loop now.")
```

## <font color="grey">Chapter 3: Functions</color>
A function is a reusable block of code that only runs when it is called.

Think of it like a recipe: you write down the steps once, and whenever you want to make that dish, you just call out the recipe's name instead of re-writing the whole instructions.

Functions help organize your code, reduce repetition, and make your programs much easier to read and maintain.

```python
# Example of a function to multipli two numbers
def product(x, y):
    mult = x*y
    return mult

# Calling a function and then printing the result
answer = product(4,6)
print(answer)
```

### Docstring
A docstring (short for documentation string) is a special kind of string used in Python to document a specific segment of code, such as a function, class, or module.

Unlike regular comments (which start with a #), docstrings are meant to explain what the code does and how to use it, rather than how it works internally.

```python
# Example of a function to multipli two numbers
def product(x, y):
    
    """ This function takes two numbers as input
    and returns their product. It cannot take 
    more than two inputs."""
    
    mult = x*y
    return mult

# Calling a function docstring
help(product)
```

## <font color="grey">Chapter 4: Classes and Objects</color>

In the real world, you know what a "car" is. You know that cars generally have wheels, a color, and a brand, and they can drive or brake. However, "car" is just a concept or a blueprint. The actual, physical car parked in your driveway is a specific instance of that concept.

In Python, that blueprint is a Class, and the actual physical car is an Object.

**A class is a user-defined blueprint or template used to create objects. It defines a set of attributes (data) and methods (functions) that the created objects will have.**

**An object is also called an <u>instance</u> of a class**


### Attributes
Attributes are just variables attached to an object. They store information that describes the object's current state.

For example, if you have a User class for a website, the attributes for a specific user object might be their <code>username</code>, <code>email_address</code>, and <code>login_count</code>.


### Methods
Methods are functions that belong to a specific object. The only major difference between a regular Python function and a method is that a method is defined inside a class, and its first parameter is almost always <code>self</code>.

Why <code>self</code>? Because a method needs a way to know which specific object is calling it so it can look up or change that object's attributes.

```python
# Example

# Superhero is a class
class Superhero:
    def __init__(self, name, superpower):
        # name, superpower and health are 'Attributes of the Superhero class
        self.name = name
        self.superpower = superpower
        self.health = 100

    # This is a method
    def take_damage(self, amount):
        self.health -= amount
        print(f"{self.name} took {amount} damage! Health is now {self.health}.")

# This is a class object or instance
myHero = Superhero('Batman', 'No Superpower')

# This is how we use a method on a class instance
print(myHero.health)
myHero.take_damage(30)
print(myHero.health)
# Output: Batman took 30 damage! Health is now 70.
```
