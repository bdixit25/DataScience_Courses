# <font color='blue'>Introduction to Data Visualisation</font>

Data visualization is the graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.    

In the world of Big Data, data visualization tools and technologies are essential to analyze massive amounts of information and make data-driven decisions.


## Why Data Visualisation in Important

### 1. Ease of Processing Visual Information
The human brain is wired to process visual information much faster than text or raw numbers. It is significantly easier to spot a trend in a line graph than it is to look at a spreadsheet with thousands of rows.

### 2. Easier to Identify Patterns and Trends
Large datasets can be incredibly overwhelming. Visualization condenses complex information, allowing you to quickly identify relationships, correlations, and shifts over time that might otherwise remain hidden.

### 3. Outliers and Anomalies
When you plot data points visually, errors in data collection, unexpected spikes, or unusual customer behavior instantly stand out as points sitting far away from the rest of the cluster.

### 4. Storytelling
Whether you are presenting to stakeholders, executives, or non-technical team members, charts tell a story that anyone can digest. It bridges the gap between data science and business strategy.

## Common Types of Visualizations & When to Use Them

| Visualization Type | Best Used For | Example Case |
| :--- | :--- | :--- |
| **Line Chart** | Tracking changes or trends over time | Monthly car sales revenue over a year |
| **Bar Chart** | Comparing quantities across different categories | Total sales by car body style (Sedan vs. SUV) |
| **Scatter Plot** | Finding relationships/correlations between two variables | Car base price vs. Actual sold price |
| **Histogram** | Showing the distribution of a continuous dataset | The age distribution of dealership buyers |

---

## The Core Components of a Great Visualization

Whenever you build a visualization in Python (using libraries like `matplotlib` or `seaborn`), you should always ensure it contains these essential structural elements:

* **Clear Title:** Tells the viewer exactly what they are looking at.
* **Axis Labels:** Explicitly names the variables on the X and Y axes, including units of measurement (e.g., Price in USD, Time in Months).
* **Legend:** Explains what different colors, shapes, or line styles represent if multiple categories are plotted together.
* **Clean Layout:** Avoids unnecessary clutter, gridlines, or 3D effects that distract from the data itself.

## <font color='blue'> Line Plot</font>


A **line plot** (or line chart) is a type of data visualization that displays information as a series of data points called 'markers' connected by straight line segments. It is one of the most basic and commonly used charts in data science.

---

### Key Use Case Scenarios

Line plots are ideal for specific analytical situations:

#### 1. Tracking Changes Over Time (Time-Series Analysis)
This is the most common use case. When the X-axis represents chronological intervals (seconds, days, months, or years), a line plot perfectly illustrates the acceleration, deceleration, or stability of a metric.
* **Example:** Tracking a company's stock price fluctuations over a fiscal year.
* **Example:** Monitoring website traffic dips and spikes during different hours of the day.

#### 2. Identifying Long-Term Trends
Because the human eye easily follows a continuous line, these plots are excellent for filtering out short-term noise to spot overarching upward or downward trajectories.
* **Example:** Analyzing global temperature anomalies over the last century to observe climate change patterns.

#### 3. Comparing Multiple Categories Over the Same Period
By plotting multiple colored lines on the same axes, you can quickly contrast how different groups perform under the exact same timeline conditions.
* **Example:** Comparing the unit sales growth of iPhone vs. Samsung Galaxy vs. Google Pixel over a 5-year span.

#### 4. Continuous Variable Relationships
While scatter plots are typically used for relationships, a line plot works well if the independent variable on the X-axis is continuous and ordered, allowing you to see the step-by-step progression of a dependent variable.
* **Example:** Plotting the compound interest growth of a savings account relative to the number of years invested.

---

### When NOT to Use a Line Plot

* **Categorical Data:** If your X-axis consists of unordered categories (e.g., Car Brands like *Toyota, Ford, BMW*), a **Bar Chart** is a much better choice. Connecting these brands with a line implies a continuous sequence or relationship that does not exist.
* **Too Many Categories:** Plotting more than 4 or 5 lines on a single chart creates a chaotic "spaghetti plot" that is completely unreadable.

### <font color='pink'>Example</font>

```python
# plotting vehicle sale data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataset for practice 
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sedans_Sold': [45, 38, 52, 60, 65, 58, 42, 40, 48, 55, 62, 70],
    'SUVs_Sold': [20, 25, 32, 40, 55, 68, 72, 75, 60, 52, 48, 44]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
# plotting using plt

plt.plot(df['Month'], df['Sedans_Sold'], label = 'Sedans')
plt.plot(df['Month'],  df['SUVs_Sold'], label = 'SUVs')
plt.title('Car Sale Trends (2026)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Month')
plt.ylabel('Cars Sold')
plt.legend()
plt.show()
```

```python
# plotting using seaborn

plt1 = sns.lineplot(data=df, x=df['Month'], y=df['Sedans_Sold'], label='Sedans')
plt2 = sns.lineplot(data=df, x=df['Month'], y=df['SUVs_Sold'], label='SUVs')

plt2.set(title="Car Sale Trends (2026)", xlabel="Month", ylabel="Units Sold")
```

## <font color='blue'> Bar Plot</font>

A **bar plot** (or bar chart) represents categorical data with rectangular bars, where the length or height of each bar is proportional to the values they represent. The bars can be plotted vertically or horizontally.

One axis shows the specific categories being compared, while the other axis represents a discrete or measured value (such as counts, sums, or averages).



---

### Key Use Case Scenarios

Bar plots excel in situations where you need to compare distinct, independent categories:

#### 1. Comparing Quantities Across Categories
This is the baseline use case. If you have distinct groups and want to see which one has the highest or lowest metric value, a bar chart makes the differences instantly obvious.
* **Example:** Comparing the total number of cars sold by different manufacturers (*Toyota vs. Honda vs. Ford*).
* **Example:** Visualizing quarterly revenue targets across different regional offices.

#### 2. Displaying Frequency Distributions of Categorical Data
When you want to see how data breaks down structurally into defined bins or groupings.
* **Example:** Counting how many cars in your dealership lot belong to specific body styles (*Sedan, SUV, Truck, Coupe*).
* **Example:** Displaying the number of customer support tickets categorized by issue type (*Billing, Login, Bug*).

#### 3. Comparing Part-to-Whole Relationships (Stacked Bar Charts)
When bars are split into colored segments (stacked), you can see both the cumulative total of a category and how individual sub-categories contribute to that total.
* **Example:** A bar showing total monthly sales, segmented internally by parts vs. vehicle sales vs. service revenue.

#### 4. Grouped Category Comparisons (Clustered Bar Charts)
By placing two or more bars side-by-side under a single category label, you can easily track performance variations across sub-groups.
* **Example:** Comparing sales figures for 2025 vs. 2026 across different car models.

---

### Horizontal Bar Plots: A Useful Alternative

When category names are exceptionally long (e.g., *"Electric Vehicle Battery Infrastructure"*), vertical bar charts force text to tilt diagonally, making it hard to read. Flipping the chart horizontally keeps your text aligned naturally from left to right.



---

### When NOT to Use a Bar Plot

* **Continuous Time-Series Trends:** If you want to show a continuous metric moving over time (like stock prices changing minute-by-minute), use a **Line Plot**. A bar chart chops time into chunky blocks, which interrupts the visual flow of a trend line.
* **Too Many Categories:** If your dataset contains 50 different individual categories, a bar chart becomes overcrowded, resulting in a wall of unreadable text and thin sticks. Consider grouping smaller categories into an *"Other"* bucket first.

## <font color='blue'> Bar Plot</font>

A **scatter plot** (also known as a scatter chart or scatter graph) is a type of data visualization that uses dots to represent the values obtained for two different numeric variables. The position of each dot on the horizontal (X) and vertical (Y) axes indicates values for an individual data point. 

Scatter plots are primarily used to observe and show **relationships** and **correlations** between two continuous variables.



---

### Key Use Case Scenarios

Scatter plots are the go-to choice for exploratory data analysis when dealing with paired numerical data:

#### 1. Detecting Correlations and Relationships
The primary use case is to see if one variable changes when another variable changes. This relationship can be:
* **Positive Correlation:** As X increases, Y increases (dots trend upward from left to right).
* **Negative Correlation:** As X increases, Y decreases (dots trend downward from left to right).
* **No Correlation:** The dots are randomly scattered with no visible pattern or slope.

* **Example:** Plotting a car's engine horsepower (X) against its fuel efficiency/MPG (Y) to see if more powerful cars consume more fuel.

#### 2. Identifying Clusters and Groupings
When you look at a scatter plot, you might notice that data points naturally gather into tight groups or "clouds." This indicates distinct subpopulations within your data.
* **Example:** Plotting customer age against annual spending to find target demographics (e.g., a distinct cluster of high-spending young adults).

#### 3. Spotting Outliers and Anomalies
Because every data point is plotted individually, values that deviate drastically from the general trend stand out immediately as isolated dots far away from the rest of the pack.
* **Example:** Finding a vehicle that sold for an unusually low price given its high baseline luxury specs, signaling a massive discount or a data entry error.

#### 4. Comparing Trends Across Sub-Groups (Using Color/Hue)
By mapping a third, categorical variable to the color or shape of the dots, a scatter plot can show how a relationship changes across different categories.
* **Example:** Plotting mileage vs. selling price, while coloring the dots by whether the vehicle is a *Sedan*, *SUV*, or *Truck*.

---

### When NOT to Use a Scatter Plot

* **When Data is Purely Categorical:** If your variables consist of non-numeric categories (like *Car Make* or *Country*), a scatter plot will just create a series of vertical columns of dots. Use a **Bar Plot** instead.
* **When Looking at Continuous Time Trends:** If you want to trace a single metric's exact movement over chronological time, a **Line Plot** is superior because it structurally connects the data points to show the sequence of events.
* **Too Many Data Points (Overplotting):** If you try to plot hundreds of thousands of dots on a single chart, they will overlap into a giant, solid blob, hiding the actual density of the data.

```python
# loading the data samples

# Create the practice dataset
coffee_data = {
    'Outside_Temp_C': [4, 8, 12, 15, 19, 22, 26, 28, 31, 24, 16, 7],
    'Hot_Coffees_Sold': [280, 245, 210, 185, 140, 110, 85, 65, 50, 95, 170, 260],
    'Iced_Coffees_Sold': [20, 35, 55, 80, 130, 175, 220, 250, 290, 190, 90, 30]
}

df_coffee = pd.DataFrame(coffee_data)
print(df_coffee.head())
```

## <font color='blue'> Box Plot</font>

A **box plot** (sometimes called a box-and-whisker plot) is a standardized way of displaying the distribution of a continuous dataset based on a **five-number summary**. Instead of showing every individual data point, it condenses the data to highlight central tendency, spread, and symmetry.

The five numbers summarized in a box plot are:
1. **Minimum:** The lowest data point excluding outliers.
2. **First Quartile (Q1 / 25th Percentile):** The middle number between the smallest number and the median.
3. **Median (Q2 / 50th Percentile):** The middle value of the dataset.
4. **Third Quartile (Q3 / 75th Percentile):** The middle value between the median and the highest value.
5. **Maximum:** The highest data point excluding outliers.



---

### Key Use Case Scenarios

Box plots are incredibly powerful for exploratory data analysis, specifically in these scenarios:

#### 1. Comparing Distributions Across Multiple Groups
This is where box plots truly shine. By placing multiple box plots side-by-side, you can instantly compare the median, range, and spread of several categories simultaneously.
* **Example:** Comparing the daily revenue distributions across different branches of a restaurant chain to see which location is most consistent.
* **Example:** Comparing starting salaries based on different college majors.

#### 2. Spotting Outliers and Anomalies
Box plots use a mathematical formula (the Interquartile Range or $IQR = Q3 - Q1$) to identify data points that fall abnormally far from the rest of the data. These are plotted as isolated dots beyond the "whiskers."
* **Example:** Identifying abnormally high transactions in credit card data to screen for potential fraud.

#### 3. Checking Data Symmetry and Skewness
The position of the median line inside the box shows you whether your data is balanced or skewed.
* **Left-Skewed:** If the median line is closer to the top/right of the box.
* **Right-Skewed:** If the median line is closer to the bottom/left of the box.
* **Symmetric:** If the median line sits perfectly in the middle.

#### 4. Visualizing Data Spread and Variability
The height (or width) of the box represents the middle 50% of your data. A tall box means high variability and volatility, while a short box means the data points are highly consistent and tightly grouped.

---

### When NOT to Use a Box Plot

* **When the Shape of the Distribution Matters (Multimodal Data):** Box plots summarize data into boxes, meaning they can completely hide the structure of data with multiple peaks (bimodal or multimodal distributions). If you want to see the exact shape of a distribution curve, a **Histogram** or **Density Plot** is much better.
* **Small Datasets:** If you only have 5 or 10 data points, breaking them into quartiles isn't mathematically meaningful. Just plot the raw points directly on a scatter or strip plot.

```python
# dataset for practising plotting box plots

delivery_data = {
    'Restaurant_Type': ['Pizza', 'Pizza', 'Pizza', 'Pizza', 'Sushi', 'Sushi', 'Sushi', 'Sushi', 
                       'Burgers', 'Burgers', 'Burgers', 'Burgers', 'Tacos', 'Tacos', 'Tacos', 'Tacos'],
    'Delivery_Time_Min': [25, 30, 28, 55, 40, 45, 42, 38, 15, 18, 22, 12, 20, 25, 23, 48]
}

df_delivery = pd.DataFrame(delivery_data)
print(df_delivery.head())
```

```python
# plotting using matplotlib

plots = []
labels = []

#for food_item in df_delivery['Restaurant_Type']:
for set_food_item in set(df_delivery['Restaurant_Type']):
    this_item_time = []
    for food_item, dlv_time in zip(df_delivery['Restaurant_Type'], df_delivery['Delivery_Time_Min']):
        if food_item == set_food_item:
            this_item_time.append(dlv_time)

    plots.append(this_item_time)
    labels.append(set_food_item)

plt.boxplot(plots, tick_labels=labels)
plt.title('Food Delivery Times')
plt.show()

```

```python
# Using seaborn

sns.boxplot(data=df_delivery, x=df_delivery['Restaurant_Type'], y=df_delivery['Delivery_Time_Min'])

plt.xlabel('Restaurant Type')
plt.ylabel('Avg. Delivery Time')
plt.title('Avg. Delivery Time vs Restaurant Type', fontsize=14, fontweight='bold', pad=15)
```

## <font color='blue'>Histograms</font>

A **histogram** is a data visualization that shows the shape of the distribution of a single, continuous numerical variable. It works by splitting the data into consecutive, non-overlapping intervals called **bins** (or buckets) along the horizontal (X) axis, and then drawing a bar over each bin to show the **count** (or frequency) of data points that fall into it.

While it looks visually similar to a bar chart, a histogram maps continuous numeric intervals on *both* axes, meaning the bars always touch each other to represent an unbroken spectrum of data.



---

### Key Use Case Scenarios

Histograms are essential during the initial phase of data analysis to understand the underlying nature of your data:

#### 1. Identifying the Distribution Shape (Normal, Skewed, etc.)
The most common use case is to see how your data clusters around a center point. You can instantly spot patterns like:
* **Symmetric/Normal Distribution:** A classic "bell curve" where most data clusters around the exact center, tapering off evenly on both sides.
* **Right-Skewed (Positive):** A long tail stretching off to the right, showing that while most values are low, a few exceptionally high values exist.
* **Left-Skewed (Negative):** A long tail stretching off to the left.

* **Example:** Plotting the exam scores of a large university class to see if the test was too hard (right-skewed) or perfectly balanced (bell curve).

#### 2. Discovering Bimodal or Multimodal Patterns
Unlike box plots, histograms will clearly show you if your data has multiple peaks. If a histogram has two separate "humps," it means you are likely dealing with two completely different sub-groups mixed into one dataset.
* **Example:** Plotting the restaurant's daily order times. You will likely see a bimodal histogram with one huge peak around lunch (12 PM) and another around dinner (6 PM), with a valley in between.

#### 3. Understanding Data Spread and Range
A histogram tells you exactly where your data begins and ends, and where the densest clusters live.
* **Example:** Plotting website loading speeds (in seconds) to check if the vast majority of your users are getting a sub-2-second experience or if there is a wide, chaotic spread.

#### 4. Detecting Outliers and Gaps
If there is a blank space on your X-axis followed by an isolated single bar far to the right, it instantly signals an extreme value gap in your collection.

---

### The Core Difference: Histogram vs. Bar Plot

| Feature | Histogram | Bar Plot |
| :--- | :--- | :--- |
| **X-Axis Type** | **Continuous Numerical** (e.g., Age 0-100, Income $0-$1M) | **Categorical / Discrete** (e.g., Apple, Banana, Orange) |
| **Spacing** | Bars **must touch** (represents a continuous range) | Bars have **spaces between them** (distinct groups) |
| **Ordering** | Must stay in strict chronological/numerical order | Can be rearranged freely (e.g., alphabetical, highest to lowest) |

```python
# plotting a Gaussian distribution

import numpy as np

random_numbers = np.random.normal(loc=5, scale=2, size=10000) #mean=5, standard_deviation=2, total entries=10000
plt.hist(random_numbers, bins=50, edgecolor='white')
```

```python
# Using seaborn
sns.histplot(random_numbers, binwidth=0.5)

plt.title('Gaussian Distribution', fontsize=15, fontweight='bold', pad=14)
plt.xlabel('Random Gaussian Numbers')
plt.ylabel('Number of Entries')
plt.show()
```

## <font color='blue'>What is a Pie Chart?</font>

A **pie chart** is a circular data visualization tool that displays information by dividing a circle into distinct "slices." Each slice represents a specific category, and its size—specifically its angle and arc length—is directly proportional to the quantity or percentage it represents relative to the entire dataset. 

Think of it like an actual pie: the entire pie represents 100% of your data, and each slice is a specific, proportional piece of that total.

---

### Use Case Scenarios

Pie charts are most effective when you need to provide a quick, high-level overview of proportional data. Here are the most common scenarios where they are the right tool for the job:

* **Displaying Parts of a Whole:** This is their primary function. They are excellent for illustrating how a single total is broken down into percentages, such as the market share of different smartphone brands or a household's monthly budget allocation.
* **Highlighting a Dominant Category:** If one category makes up the vast majority of the data (e.g., one slice is 75% while the rest are tiny slivers), a pie chart makes that contrast immediately visually striking to the viewer.
* **Simplifying Survey Results:** They are great for visualizing simple, categorical survey responses with limited options, such as "Yes / No / Undecided."
* **Comparing a Small Number of Categories:** Pie charts work best when comparing a very limited dataset—ideally between two to five distinct categories. 

### When to Avoid Pie Charts

To give you the most accurate overview, it is just as important to know when *not* to use a pie chart:

* **Too Many Categories:** If you have more than five or six slices, the chart quickly becomes cluttered, difficult to read, and loses its visual impact. The smaller slices become impossible to distinguish.
* **Comparing Similar Values:** The human eye struggles to accurately judge slight differences in angles or 2D areas. If you need to show that Category A is 23% and Category B is 25%, a **bar chart** is much more effective because humans are better at comparing straight lengths.
* **Showing Trends Over Time:** Pie charts only represent a single, static snapshot in time. To show how data changes over months or years, a **line graph** is the standard and correct choice.

## <font color='blue'> What is a Bubble Plot? </font>

A **bubble plot** (or bubble chart) is a variation of a scatter plot that allows you to visualize **three or four dimensions** of data simultaneously on a two-dimensional grid. 

While a standard scatter plot uses $x$ and $y$ coordinates to map two variables, a bubble plot adds a third variable by varying the **size (area)** of the data points (the "bubbles"). Frequently, a fourth variable is introduced by assigning different **colors** to the bubbles, typically representing distinct categories or a continuous gradient scale.

---

### Use Case Scenarios

Bubble plots are highly effective for displaying complex, multi-variable relationships without requiring a 3D graph. Here are the most common scenarios where they excel:

* **Economic and Demographic Comparisons:** They are famously used to compare countries or regions. For example, mapping **GDP per Capita** ($x$-axis) against **Life Expectancy** ($y$-axis), where the **size of the bubble** represents the country's population, and the **color** represents the continent. 
* **Business and Project Portfolio Management:** Companies use them to evaluate projects or products. A chart might map **Risk** ($x$-axis) vs. **Expected ROI** ($y$-axis), with the **bubble size** representing the project's total cost, helping executives visually identify low-risk, high-reward, budget-friendly initiatives.
* **Website and Digital Marketing Analytics:** You can evaluate user behavior by plotting **Bounce Rate** ($x$-axis) vs. **Conversion Rate** ($y$-axis), where the **bubble size** represents total traffic volume (page views) and the **color** represents the marketing channel (Organic, Paid, Social).
* **Scientific and Environmental Data:** Researchers use them to show environmental relationships, such as plotting **Temperature** ($x$-axis) vs. **Humidity** ($y$-axis), with the **bubble size** indicating rainfall volume.

---

### When to Avoid Bubble Plots

Because bubble plots pack a lot of information into one visual, they can easily become ineffective if misused:

* **Too Many Data Points:** If your dataset has hundreds of rows, the bubbles will overlap significantly, creating a cluttered "cloud" where data points hide behind one another.
* **Comparing Exact Sizes:** The human brain is not very good at accurately judging the relative area of circles. If it is critical for your audience to know the exact numeric value of the third variable, a **data table** or a **bar chart** is a safer choice.
* **Negative or Zero Values:** Bubble size relies on area. You cannot visually represent a "negative" or "zero" area size in a way that makes intuitive sense to a viewer. If your third variable contains negative numbers, a bubble plot will not work.

## <font color='blue'> What is an Area Plot? </font>

An **area plot** (or area chart) is a data visualization that displays quantitatively tracked variables over time. It is based directly on the standard line graph, but with one key difference: the space between the horizontal axis and the plotted line is completely filled in with color, texture, or shading. 

By filling in the area beneath the line, this plot draws visual emphasis to the **volume or magnitude** of the data over time, rather than just the trend or rate of change. When multiple data series are tracked together, they are typically stacked on top of each other (a *stacked area chart*), showing how much each category contributes to the changing total over time.

---

### Use Case Scenarios

Area plots are ideal when you want to show both a total trend and the individual breakdown of that total simultaneously. Here are the best scenarios for using them:

* **Tracking Total Volume with Category Breakdowns:** They excel at showing cumulative growth alongside individual contributions. For example, plotting **Total Company Revenue** ($y$-axis) over **Years** ($x$-axis), with the area split into stacked colors representing different product lines.
* **Visualizing Part-to-Whole Relationships Over Time:** If you want to see how the market share of different competitors has shifted over a decade, a 100% stacked area chart scales the vertical axis to 100% and shows how the proportional slice of each company grows or shrinks over time.
* **Highlighting Broad Trends and Magnitudes:** Because the filled-in area creates a heavy visual weight, they are great for public-facing reports or dashboards where you want the audience to instantly grasp a massive surge or drop in volume (e.g., website traffic volume over a holiday weekend).

---

### When to Avoid Area Plots

Area plots can easily become misleading or unreadable if applied to the wrong data:

* **Too Many Overlapping Categories:** If you use a basic, unstacked area plot with more than two or three categories, the filled-in shapes will overlap and hide each other. This is known as "occlusion." 
* **Comparing Exact Values Across Categories:** In a stacked area chart, every layer after the bottom one sits on an uneven foundation. Because the baseline for the upper categories is constantly shifting up and down, it is very difficult for the human eye to accurately judge the exact value or individual trend of a middle layer. If tracking individual trends precisely is your goal, use a standard **line graph** instead.
* **Fluctuating or Non-Continuous Data:** Area plots rely on a continuous, connected sequence of data (like time). ## <font color='blue'> What is a Treemap? </font>

A **treemap** is a data visualization technique used to display hierarchical (nested) data using nested rectangles. 

Think of it as a way to see a whole system broken down into its component parts all at once. The chart is structured like this:
* **The Frame:** The entire plot area represents the total value of the main dataset (100%).
* **The Branches & Leaves:** The main area is divided into larger rectangles (branches/categories), which are then subdivided into smaller rectangles (leaves/subcategories).
* **The Dimensions:** * **Size:** The area of each rectangle is proportional to a specific quantitative value (e.g., total sales or file size).
  * **Color:** Often used to represent a second metric, such as growth rate, profit margins, or distinct category types.

---

### Use Case Scenarios

Treemaps excel at displaying hundreds of categories simultaneously, allowing users to spot patterns, anomalies, and proportions at a single glance.



#### 1. Financial Market Analysis (The "Stock Market Map")
This is perhaps the most famous use case. Financial platforms use treemaps to show the performance of an entire stock index (like the S&P 500) at once. The rectangles are grouped by sectors (Tech, Healthcare, Energy), the size of each rectangle represents the company's market cap, and the color represents daily price movement (Green for gains, Red for losses).

#### 2. Disk Space and File Management
If you've ever used disk cleanup software (like WinDirStat or GrandPerspective), you've looked at a treemap. It maps out your entire hard drive, where larger rectangles represent massive files or folders, letting you instantly pinpoint exactly what is eating up your storage space.

#### 3. E-commerce and Retail Inventory
Retailers use treemaps to analyze sales across complex product hierarchies. For example, a clothing brand can visualize sales across categories (Men's, Women's, Kids) and subcategories (Shirts, Pants, Outerwear) to see which specific items are driving revenue.

#### 4. Website Traffic and Analytics
Digital marketers use them to understand where traffic is coming from. The map can display traffic sources grouped by channels (Organic, Paid, Social) and split down to specific landing pages, with sizes representing page views and colors representing bounce rates.

---

### When to Avoid Treemaps

While incredibly efficient, treemaps are not always the right choice:

* **Comparing Non-Hierarchical Data:** If your data doesn't have categories and subcategories, a standard **bar chart** is cleaner and easier to read.
* **Precise Value Judgments:** Just like pie charts, the human eye struggles to accurately compare the exact areas of different rectangles, especially when their aspect ratios (width-to-height) vary. If exact number comparisons matter most, a **data table** is better.
* **Extreme Data Variances:** If your dataset contains a few massive categories and hundreds of microscopic ones, the tiny categories will become unreadable "slivers" that clutter the chart without adding value.

## <font color='blue'> What is a Waffle Chart? </font>

A **waffle chart** (also known as a square pie chart or grid plot) is a data visualization tool used to display progress toward a target or proportional distributions. It consists of a grid of 100 square cells arranged in a 10x10 layout. 

Each individual square represents **1% of the total dataset**. The cells are colored based on the proportions of the categories being measured, making it look strikingly similar to a breakfast waffle.



---

### Use Case Scenarios

Waffle charts are an increasingly popular alternative to pie charts because they are much easier for the human eye to read and compare. Here are the best scenarios for using them:

#### 1. Tracking Progress Toward a Goal (KPIs)
They are excellent for business dashboards tracking a single metric against a target. For example, if a fundraising campaign has achieved 74% of its goal, a waffle chart will color in exactly 74 squares, providing a clear, motivating visual of how close they are to completion.

#### 2. Highlighting Small Percentages
In a pie chart, a category that represents 1% or 2% of the data turns into an almost invisible sliver. In a waffle chart, that same data point occupies 1 or 2 distinct, cleanly bounded squares, ensuring small but critical metrics aren't overlooked.

#### 3. Displaying Human or Demographic Data
Waffle charts are heavily used in journalism and infographics because the individual squares can easily represent distinct entities (e.g., "1 square = 1 out of 100 people surveyed"). This makes abstract percentages feel more tangible to a general audience.

#### 4. Simple Category Breakdowns
Just like pie charts, they work beautifully for comparing a small number of discrete categories (ideally 2 to 4) that combine to form a whole, such as the gender breakdown of a company's workforce or the market share of major operating systems.

---

### When to Avoid Waffle Charts

Despite their clean look, waffle charts should be avoided in these situations:

* **Large or Continuous Datasets:** A waffle chart requires discrete, categorical data that can be converted into percentages. It cannot display continuous fluctuations over time or complex scientific measurements.
* **Too Many Categories:** If you have 10 different categories, your grid will become a confusing mosaic of 10 colors, destroying the chart's primary benefit of instant readability.
* **Data with Extreme Decimals:** Because a standard waffle chart relies on 100 squares, numbers are rounded to the nearest whole percentage. If the difference between 44.2% and 44.4% is critical to your analysis, a waffle chart is too imprecise.
