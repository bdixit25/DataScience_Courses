## What are Dashboards?

In the context of data analysis and business intelligence, a **dashboard** is a visual interface that provides an at-a-glance view of key performance indicators (KPIs), metrics, and critical data points relevant to a particular objective or business process.

> Think of a dashboard like the instrument panel in a car: it doesn't show you the mechanical details of the engine, but it provides immediate, actionable information (speed, fuel levels, engine temperature) necessary to drive the vehicle safely and efficiently.

---

## Core Characteristics of a Dashboard

To be considered a true dashboard, the interface generally meets four criteria:

* **Visual Representation:** It uses charts, graphs, gauges, and maps to make complex data sets easier to comprehend quickly.
* **Centralized View:** It consolidates data from multiple sources into a single screen or page.
* **Real-time or Near Real-time:** It is designed to reflect current conditions so users can make timely decisions.
* **Actionability:** It highlights trends or anomalies that prompt the user to investigate further or take specific actions.

---

## Common Types of Dashboards

Dashboards are typically categorized by their purpose and the audience they serve:

| Dashboard Type | Primary Purpose | Intended Audience |
| :--- | :--- | :--- |
| **Operational** | Monitors day-to-day processes and performance. | Managers, team leads, floor staff. |
| **Strategic** | Tracks long-term goals and organizational health. | Executives, C-suite, board members. |
| **Analytical** | Explores data to find patterns, correlations, or root causes. | Data analysts, scientists, researchers. |

---

## Why Are They Useful?

* **Efficiency:** They eliminate the need to manually aggregate data from spreadsheets or multiple software platforms.
* **Context:** They allow users to see how individual metrics relate to broader goals.
* **Accountability:** They make performance visible, which can encourage focus on specific objectives.
* **Trend Identification:** They make it much easier to spot anomalies, growth, or decline over time compared to reading raw data logs.

---

## Key Components

* **Data Sources:** The underlying databases, APIs, or files feeding the dashboard (e.g., Salesforce, Google Analytics, SQL databases).
* **Widgets/Visualizations:** The individual charts (bar charts, line graphs, heat maps) that display specific metrics.
* **Filters:** Tools that allow users to change the view based on time ranges, regions, or specific departments.

## <font color='blue'> How to use Ploty to create DashBoards? </font>

### 1. The Core Components

* **Plotly (The Engine):** A library used to create the actual charts (scatter plots, bar charts, heatmaps). It is interactive, meaning users can zoom, pan, and hover over data points.
* **Dash (The Interface):** A framework built on top of Plotly, React.js, and Flask. It allows you to build web applications (dashboards) entirely in Python, without needing to know HTML, CSS, or JavaScript.

---

### 2. Basic Workflow for a Dashboard

To create a dashboard, you generally follow this structure:

* **Layout:** Define the visual arrangement using `dash_html_components` (for standard HTML like Divs and Headers) and `dash_core_components` (for interactive elements like dropdowns and graphs).
* **Callbacks:** These are Python functions that trigger automatically whenever an input (like a slider or dropdown) changes. The callback takes the input, processes the data, and updates the output (the graph).

# <font color='orange'> Introduction to Dash </font>

Dash is an open-source Python framework specifically designed for building analytical web applications. It is the go-to tool for data scientists and analysts who want to turn their data models and visualizations into interactive, professional-looking dashboards without needing to learn HTML, CSS, or JavaScript.

---

### Why Dash is Unique

Most web development requires expertise in multiple languages (HTML for structure, CSS for style, JavaScript for interactivity, and a backend language like Python for data processing). **Dash simplifies this by allowing you to handle the entire stack using only Python.**

---

### The Core "Stack" Under the Hood

Dash acts as a wrapper around three powerful technologies, letting you control them via simple Python code:

* **Flask:** Acts as the web server, handling the communication between your app and the browser.
* **Plotly.js:** Provides the high-quality, interactive charting engine for the graphs you see on screen.
* **React.js:** Renders the user interface components (like dropdowns, sliders, and buttons) in the browser.

---

### The 3-Step Building Process

Building a Dash app is usually broken down into these three consistent steps:

1. **Define the Layout:** Use Python modules (`dash.html` and `dash.dcc`) to describe how your app looks. Instead of writing HTML tags, you write Python functions that represent those tags.
2. **Write Callbacks:** This is the "brain" of the app. Use a special decorator (`@app.callback`) to link user inputs (like a dropdown menu) to output elements (like a graph). When the input changes, the callback function automatically triggers and refreshes the graph.
3. **Run the App:** Call `app.run()` to start the built-in development server, which hosts your app locally so you can interact with it in your browser.

---

### Key Advantages

* **Interactivity:** Because it’s built on Plotly, users can hover, zoom, and filter data directly within the dashboard.
* **Rapid Prototyping:** You can go from a static data analysis script to a fully functional, shareable web app in just a few dozen lines of code.
* **Extensibility:** While it’s "pure Python," it is highly customizable. If you do know CSS, you can add your own styles; if you do know JavaScript, you can build your own custom components.
* **AI-Friendly:** Modern AI coding agents are highly efficient at generating Dash apps because the framework’s declarative, structure-based syntax is very predictable for LLMs.

---

### When should you use it?

* **When you need more than a static chart:** If you need filters, toggles, or dynamic data updates, Dash is the right tool.
* **When you work in Python:** If your data cleaning and modeling are already done in Pandas or NumPy, Dash integrates with those libraries natively.
* **When you need a quick internal tool:** It is perfect for turning a Jupyter Notebook analysis into a tool that a non-technical colleague can use.

### 1. Dash Layouts: The Component Tree

Dash layouts are built as a **tree of components**. You start with a parent container and "nest" children inside it.

* **The Root:** Usually a top-level container (like `html.Div`) that holds everything.
* **The Branches:** You can place rows, columns, or other structural elements inside that parent.
* **The Leaves:** These are the final display elements, such as a graph (`dcc.Graph`), a piece of text (`html.H1`), or an input field (`dcc.Input`).

Because it is a tree structure, when you define `app.layout` in your Python code, you are essentially defining the structure of the webpage as a nested list of Python objects.

---

### 2. Popular Modules

To build this tree, Dash provides two primary libraries (modules) that act as your building blocks:

#### Dash HTML Components (`dash.html`)
This module provides Python classes for every standard HTML tag. It handles the structure and appearance (CSS) of your app.

* `html.H1("Title")`: Creates an `<h1>` tag. H2, H3, H4, H5 etc are also section headings.
* `P`: Creates a paragraph
* `Br`: Line Break
* `A`: Anchor
* `html.Div(...)`: Creates a `<div>` tag, which is the most common way to group components.
* `html.Button("Click Me")`: Creates a `<button>` tag.

#### Dash Core Components (`dash.dcc`)
This module provides high-level interactive components. These handle user input and data visualization and contain built-in "intelligence" (like state management for dropdowns).

* `dcc.Graph()`: Used to display your Plotly charts.
* `dcc.Dropdown()`: Creates a selectable menu.
* `dcc.RadioItems()`: Radio Items (only one item can be selected)
* `dcc.Input()`: Text input boxes
* `dcc.Slider()`: Creates an interactive range slider.
* `dcc.RangeSlider()`: Interactive range slider.
* `dcc.DatePicker()`: Adds a calendar selector.


### <font color='blue'> Example </font>
<code>
from dash import html, dcc

app.layout = html.Div([
    html.H1("My Dashboard"),                # HTML component for styling
    dcc.Dropdown(options=['A', 'B']),       # Core component for interaction
    dcc.Graph(id='my-graph')                # Core component for data
])
</code>

In this code, the `html.Div` is the "parent" of the tree, and the `H1`, `Dropdown`, and `Graph` are the "children" nested inside it.

```python
# Let's create a dashboard to view this data

from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px

df=pd.read_csv('online_courses_harvard_mit.csv')

app=Dash()

df = df[df['Participants (Course Content Accessed)'] >= 20000]

app.layout=html.Div([
    html.H1('Course Participants'),
    html.P(['This is an example dashboard for the rankings of imaginary universities.',
            html.Br(),
            html.A('To go Google', href='www.google.com')]),
    dcc.RadioItems(id='subject-radio', options=df['Course Subject'].unique(), value='Computer Science'),
    dcc.Dropdown(id='uni-dropdown', options=df['Institution'].unique(), value='MITx'),
    dcc.Graph(id='pie-chart')  
])

@app.callback(
    Output('pie-chart', 'figure'),
    [Input('subject-radio', 'value'),
     Input('uni-dropdown', 'value')]
)

def update_chart(selected_subject, selected_uni):

    df_selected = df[(df['Course Subject'] == selected_subject) & (df['Institution'] == selected_uni)]
    df_grouped = df_selected.groupby('Course Title')['Participants (Course Content Accessed)'].sum().reset_index()

    figure=px.pie(
            df_grouped,
            names='Course Title',
            values='Participants (Course Content Accessed)'
        )
    return figure

if __name__ == '__main__':
    app.run(debug=True)
```
