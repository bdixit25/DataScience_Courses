## Practice Numpy and Pandas with datasets

```python
# generating the datasets

import pandas as pd
import numpy as np

# Seed for reproducible randomness
np.random.seed(42)
n_records = 1000

# Continuous raw data setup
years_list = list(range(2020, 2026)) # 2020 to 2025
raw_sales = np.random.randint(5000, 75000, size=(n_records, len(years_list)))

data = {
    'UniqID': [f'TRX-{1000 + i}' for i in range(n_records)],
    'ProdGroup': np.random.choice(['Electronics', 'Apparel', 'Home & Kitchen', 'Beauty', 'Sports'], n_records),
    'MarketSegment': np.random.choice(['North America', 'Europe', 'Asia-Pacific', 'Latin America', 'MEA'], n_records),
    'Fulfillment_Type': np.random.choice(['Express', 'Standard', 'Economy'], n_records),
    'Quality_Score_Avg': np.random.uniform(3.5, 5.0, n_records).round(1),
    'Tax_Exempt_Status': np.random.choice([True, False], n_records, p=[0.15, 0.85])
}

# Integrate multi-year continuous matrices
for idx, yr in enumerate(years_list):
    data[yr] = raw_sales[:, idx]

df_sales = pd.DataFrame(data)
print("E-Commerce Dataset with 1,000 records successfully loaded!")
print(f"Shape: {df_sales.shape}")
```

## <font color='blue'>Exercise 2: Column Renaming & Feature Generation</font>

1. **Rename the metadata columns** so they read cleanly: Change `ProdGroup` to `Category` and `MarketSegment` to `Region`.
2. **Generate a new metric column** named `Total_Revenue` that sums up all values horizontally across all year periods for each distinct index row profile.

```python
# 1. Rename the metadata columns so they read cleanly: Change 'ProdGroup' to 'Category' and 'MarketSegment' to 'Region'

df_cleaned = df_subset_sales.rename(columns={'ProdGroup':'Category', 'MarketSegment':'Region'})
df_cleaned.head(3)
```

```python
# 2. Generate a new metric column named 'Total_Revenue' that sums up all values horizontally across all year periods 
# for each distinct index row profile

years=[2020, 2021, 2022, 2023, 2024, 2025]

df_cleaned['Total_Revenue'] = df_cleaned[years].sum(axis=1)
df_cleaned.head(3)
```

## <font color='blue'>Exercise 4: Explicity Label Data Typing & Slicing Variable Generation</font>

1. **Safely transform every column header item** mapping inside your list configuration profile into explicit Python string formats (`str`).
2. **Generate a standalone list variable container** called `revenue_years` capturing historical string years from `'2020'` through to `'2025'`.
3. **Use that explicit variable list container map** to pull back a localized data view segment strictly tracking `Apparel` performance metrics during those targeted years.

```python
# 1. Safely transform every column header item mapping inside your list configuration profile into explicit Python string formats (str)

df_cleaned.columns = list(map(str, df_cleaned.columns))
df_cleaned.head(3)
```

```python
# 2. Generate a standalone list variable container called revenue_years capturing historical string years from '2020' through to '2025'

revenue_years = list(map(str, range(2020, 2026)))
print(revenue_years)
```

```python
# 3. Use that explicit variable list container map to pull back a localized data view segment strictly tracking 'Apparel' 
# performance metrics during those targeted years

apparel_history = df_cleaned.loc['Apparel', revenue_years]
print(apparel_history.head())
```
