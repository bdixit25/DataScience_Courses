## <font color='grey'> Chapter 3: String Operations </font>
A **String** is anything consiting alphanumeric, or special character contained within <u>*single*</u> or <u>*double quotes*</u>.
e.g. "skdf 353 kfdl 48**#23" or 'fdkskd5302'

---

A string is also an ordered sequence with each entry having its own index.  
The index start from 0 and goes from left to the right.  
The characters from the left can be accessed using negative integers.

### String Indexing

```python
# Accessing characters of a String using Index
name = "Gautama Buddha"
print(name[0])
print(name[1])
print(name[-1])
print(name[-2])
```

```python
# String Slicing
first_name = name[0:7]
print(first_name)

last_name = name[-6:]
print(last_name)
```

```python
# Stride (selecting alternate characters)

# e.g. select every second variable
name_alt_chars = name[::2]
print(name_alt_chars)
```

```python
# Stride + Slicing

# e.g. return every second variable but only upto index 7
name_stride_n_slice = name[0:7:2]
print(name_stride_n_slice)
```

### String Length

```python
# Print Length of a string
len(name)
```

### Concatenating a String

```python
# Adding strings
name = "Anthony"
surname = "George"
full_name = name + " " + surname
print(full_name)
```

### Escape Sequences in a String
There are three escape sequences in python strings:
- **\n** : To start a new line
- **\t** : To give a tab space
- **\\\\\\** : Input a backslash in the string

```python
# \n

print("My name is George. \n I live in London.")
```

```python
# \t

print("1 \t 2 \t 3 \t 4 \t 5")
print("2 \t 4 \t 6 \t 8 \t 10")
print("3 \t 6 \t 9 \t 12 \t 15")
print("4 \t 8 \t 12 \t 16 \t 20")
print("5 \t 10 \t 15 \t 20 \t 25")
```

```python
# \\
print("This is how you print \\ in a string.")
```

### String Methods
String methods are string modifying operations. Some common methods include:
- <span style="font-family: Courier New;">upper()</span> : Converts all characters in a string to uppercase
- <span style="font-family: Courier New;">replace(old_substring, new_substring)</span> : Replace a part of a string
- <span style="font-family: Courier New;">find()</span> : Prints out the fist index of the start of the sequence
- <span style="font-family: Courier New;">title()</span> : Changes first character of a string to upper case

```python
# \upper()
name = 'wonderland'
uppercase_name = name.upper()
print(uppercase_name)
```

```python
# replace a part of a string
new_name = name.replace('wonder', 'border')
print(new_name)
```

```python
# find
found_index = name.find("land")
print(found_index)
```

**<font color="red"> If the subsring is not found in the string the result is '-1'</font>**

```python
not_found_index = name.find("home")
print(not_found_index)
```

```python
# title method
eg_string = "liverpool: a beautiful city"
title_str = eg_string.title()
print(title_str)
```
