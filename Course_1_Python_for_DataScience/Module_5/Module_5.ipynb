## What is an API?

**API** stands for **Application Programming Interface**. 

In simple terms, an API is a software intermediary that allows two different applications to **talk to each other**. It acts as a digital bridge, delivering a request from one system to another, and then bringing the response back.

---

### 1. The Restaurant Analogy

The classic way to understand an API is to imagine you are dining at a restaurant:

* **You (The Client):** You are sitting at the table looking at a menu of choices. You want to order food, but you cannot go directly into the kitchen to get it.
* **The Kitchen (The Server):** This is the database or system that prepares your data/food.
* **The Waiter (The API):** The waiter is the API. They take your order (your **Request**), walk it over to the kitchen, tell the kitchen what to do, and then bring the food (the **Response**) back to your table.

---

### 2. Real-World Examples of APIs

You interact with APIs dozens of times a day without even realizing it:

* **"Log in with Google"**: When a website lets you log in using your Google credentials, it doesn't see your password. It uses Google’s API to securely ask, *"Is this user who they say they are?"*
* **Uber / Lyft Maps**: Uber didn't build a global mapping system from scratch. Their app uses the **Google Maps API** under the hood to handle location tracking and navigation.
* **Weather Apps**: The weather app on your phone doesn't own a satellite. Instead, it connects to a weather bureau's API to pull the latest temperature and forecast data instantly.

---

### 3. How an API Works (The Technical Flow)

When applications interact via an API, the process generally follows a structured four-step cycle:

```text
 [Your App]  ---- (1. Request via URL) ---->   [The API Backend]
                                                      |
                                              (2. Processes & Fetches)
                                                      |
                                                      v
 [Your App]  <--- (4. Data Delivered) <------   [Database / Server]

## Is Pandas an API?

**Yes, Pandas is an API, but in a different way** than the web APIs you usually hear about (like the Google Maps or Weather APIs).

There are actually two major types of APIs in computer science:

1. **Web APIs (Remote APIs):** Code that lets two different computers talk to each other over the internet using URLs (like fetching weather data or processing credit cards).
2. **Library/Software APIs (Local APIs):** Code that defines how a programmer interacts with a software library *inside* their own script.

**Pandas falls perfectly into the second category.**

---

### 1. Why Pandas is Considered an API

When you write `import pandas as pd`, Pandas exposes its "Interface" (the **I** in API) to you. It gives you a strict, predictable set of commands to interact with its highly optimized underlying C and Python code.

You do not need to know how Pandas manages computer memory or how it implements complex sorting algorithms under the hood. You just use its public API methods to get what you need:

* `pd.DataFrame()` is an API method for creating tables.
* `pd.read_csv()` is an API method for loading data.
* `df.dropna()` is an API method for cleaning missing values.

---

### 2. The Key Difference: Web API vs. Pandas API

| Feature | Web API (e.g., OpenWeather) | Pandas API |
| :--- | :--- | :--- |
| **Location** | Lives on a remote server on the internet. | Lives locally on your computer after you `pip install` it. |
| **Communication** | Uses HTTP requests over the web (URLs). | Uses standard Python method calls inside your script. |
| **Data Format** | Usually sends and receives **JSON** or XML text strings. | Processes data using **DataFrames** and **Series** directly in system memory. |

---

### 💡 Summary
Pandas **is** an API. Specifically, it is a **local Python library API** that acts as an abstraction layer—translating your clean, simple Python code into highly complex data manipulations under the hood.

## What is a REST API?

**REST** stands for **Representational State Transfer**. It is not a programming language or a piece of software, but rather a set of **architectural rules and design principles** for creating web APIs. 

When a web API obeys the rules of REST, it is called a **RESTful API**. 

REST APIs are the industry standard for how modern web applications send data back and forth over the internet using the standard web protocol: **HTTP** (Hypertext Transfer Protocol).

---

### 1. Core Principles of REST

To be considered "RESTful," an API must follow several strict design constraints. The three most fundamental are:

#### A. Client-Server Architecture
The user interface (the client, like a mobile app or a browser) and the data storage (the server) must remain completely independent. As long as they both know what format of data to send to each other, you can update the backend database without breaking the frontend app.

#### B. Statelessness
This is the most critical rule. The server **does not remember past requests**. Every single request from the client must be an independent package containing all the information, credentials, and context needed to understand and fulfill it. 

#### C. Uniform Interface (Resources)
In a REST API, everything is treated as a **Resource** (e.g., a user, a product, a photo, a tweet). Every resource must have a unique digital address called a **URI** (Uniform Resource Identifier) or **Endpoint**.

<code>
[https://api.shop.com/v1/products](https://api.shop.com/v1/products)      <-- Endpoint for ALL products
[https://api.shop.com/v1/products/102](https://api.shop.com/v1/products/102)  <-- Endpoint for a SPECIFIC product</code>


### 2. How REST APIs Work
* **The Request:** Communication begins with a client sending a **request** to a server via an **HTTP message**. This message typically includes a **JSON** file detailing the specific operation you want the service or **resource** to perform.
* **The Response:** The API processes the request and sends back a **response** to the client, also via an **HTTP message**, with the data usually packed inside a **JSON** format.

### Example: The NBA API Lab Overview
* **Objective:** You will use the `nba-api` Python library to analyze game data between the **Golden State Warriors** and the **Toronto Raptors**.
* **Key Metric:** The goal is to calculate the **point differential** for each game from the 'Warriors' perspective:
  * **Positive value (e.g., `3`):** The Golden State Warriors won the game by 3 points.
  * **Negative value (e.g., `-2`):** The Golden State Warriors lost the game by 2 points.
* **The Benefit of the API:** The `nba-api` library acts as an abstraction layer, meaning it will automatically handle complex backend details like **Endpoints** and **Authentication** for you under the hood.

````python
# Example

from nba_api.stats.static import teams
import matplotlib.pyplot as plt


# Takes list of dictionaries & combines them into a single dictionary
def my_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}

    for dict_ in list_dict:
        for key, value in dict_.items():
            out_dict[key].append(value)
    return out_dict

# Use 'get_teams()' method to get a list of dictionaries
nba_teams = teams.get_teams()

# Look at the first three elements
nba_teams[0:3]
````

````python
# Convert the dictionary into a table
import pandas as pd
dict_nba_team = my_dict(nba_teams)
df_teams=pd.DataFrame(dict_nba_team)
df_teams.head()
````

````python
# Finding the that contains the "Warriors" by using the column 'nickname'
df_warriors=df_teams[df_teams['nickname']=='Warriors']
df_warriors
````

````python
# Accessing 1st column of DataFrame
id_warriors=df_warriors[['id']].values[0][0]
id_warriors
````

## <font color='grey'> Chapter 2: Web Scraping</font> 

**Web scraping** (also called web harvesting, web extraction, or data scraping) is the automated process of using software to extract large amounts of data from websites. 

While an API provides a clean, structured pipeline to fetch data, web scraping is used when a website **does not provide an API**, forcing a program to read the website's raw visual layout and extract the data manually.

---

### 1. The Analogy: Human vs. Bot

Imagine you want to collect the prices of 500 laptops listed on an e-commerce site:

* **The Human Way:** You open a browser tab, click a laptop, look at the price, copy it, open an Excel sheet, paste it, go back, and repeat 500 times. This is slow, repetitive, and exhausting.
* **The Web Scraper Way:** You write a script that sends a request to the website, pulls down the raw code for all 500 pages in seconds, hunts through the code for the price labels, and automatically saves them cleanly into a CSV file.



---

### 2. How Web Scraping Works

A typical web scraping script written in Python (using libraries like **BeautifulSoup**, **Scrapy**, or **Selenium**) follows a 4-step lifecycle:

```text
 [Your Script]  ===> 1. Send HTTP Request ===>  [Target Website Server]
                                                       ||
 [Your Script]  <=== 2. Download Raw HTML <====  [Target Website Server]
      ||
 (3. Parse & Filter HTML)
      ||
      v
 [Clean CSV / DataFrame Data]
